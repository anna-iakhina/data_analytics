```python
#задание 4
df['year'] = df['order_purchase_timestamp'].astype("datetime64[ns]").dt.strftime('%Y')
#выводим в отдельный столбец год, чтобы смотреть данные по годам
```

```python
def weeks_in_month(year, month):
    _, days_in_month = calendar.monthrange(year, month)
    weeks = days_in_month // 7
    remainder_days = days_in_month % 7
    return weeks + remainder_days / 7
#посчитали, сколько недель в каждом месяце
```

```python
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df['weeks_in_month'] = df.apply(lambda row: weeks_in_month(row['year'], row['month']), axis=1)
```

```python
orders_per_month_week = df.query('order_status == "delivered"') \
                          .groupby(['customer_unique_id', 'weeks_in_month', 'year', 'month'], as_index=False) \
                          .agg({'order_id' : 'count'})
#смотрим, сколько покупок совершено по каждому месяцу рассматриваемого периода 
```

```python
orders_per_month_week
```

```python
orders_per_month_week['orders_per_week'] = orders_per_month_week['order_id']/orders_per_month_week['weeks_in_month']
```

```python
orders_per_month_week.sort_values('orders_per_week') 
#сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)
```

```python
df['order_purchase_date'] = df['order_purchase_timestamp'].astype("datetime64[ns]").dt.strftime('%Y-%m-%d')
#из order_purchase_timestamp вытаскивает дату совершения покупки
```

```python
first_orders = df.groupby(['customer_unique_id'], as_index=False).agg({'order_purchase_date': 'min'})
#находим дату совершения первой покупки
```

```python
first_orders
```

```python
first_orders.rename(columns= {'order_purchase_date' : 'first_order_purchase_date'}, inplace=True)
```

```python
first_orders['customer_unique_id'] = first_orders['customer_unique_id'].astype(str)
df['customer_unique_id'] = df['customer_unique_id'].astype(str)
```

```python
first_orders.reset_index(drop=True, inplace=True)
df_merged = df.merge(first_orders, how='inner', left_on='customer_unique_id', right_on='customer_unique_id')
```

```python
df_merged
#объединяем исходную табоицу с данными о первой покупке 
```

```python
df_merged['first_order_purchase_date'] = df_merged['first_order_purchase_date'].astype("datetime64[ns]")
df_merged['order_purchase_date'] = df_merged['order_purchase_date'].astype("datetime64[ns]")
```

```python
df_merged.dtypes
```

```python
df_merged['months_diff'] = (df_merged['order_purchase_date'] - df_merged['first_order_purchase_date']) / np.timedelta64 ( 1 , 'M')
#находим разницу в месяцах между датой первой покупки и датой следующей покупки
#months_diff принимаем за номер месяца когорты (то есть, если months_diff = 2, это значит, что пользователь совершил покупку на второй месяц после первой покупки)
```

```python
df_merged.head()
```

```python
df_merged['months_diff'] = df_merged['months_diff'].round()
```

```python
df_merged['order_purchase_date'] = df_merged['order_purchase_date'].astype("datetime64[ns]").dt.strftime('%Y-%m')
df_merged['first_order_purchase_date'] = df_merged['first_order_purchase_date'].astype("datetime64[ns]").dt.strftime('%Y-%m')
```

```python
df_merged.head()
```

```python
df_merged['months_diff'] = df_merged['months_diff'].astype(int)
```

```python
cohorts_by_month = df_merged.groupby(['first_order_purchase_date'], as_index=False).agg({'customer_unique_id' : 'count'})
```

```python
cohorts_by_month
#смотрим когорты по месяцу первой покупки
```

```python
cohort_data = df_merged.groupby(['first_order_purchase_date', 'months_diff'], as_index=False).agg({'customer_unique_id' : 'count'})
#формируем df для построения когортной таблицы с данными о первом месяце покупки и номером когортного месяца (months_diff)
```

```python
cohort_data
```

```python
cohort_data['customer_unique_id'] = cohort_data['customer_unique_id'].astype(int)
```

```python
cohort_table = cohort_data.pivot(index='first_order_purchase_date', columns='months_diff', values='customer_unique_id')
```

```python
cohort_table
```

```python
retention_rate = cohort_table.divide(cohort_table[0], axis=0)
```

```python
retention_rate
```

```python
ur_style = (retention_rate
            .style
            .set_caption('User retention by cohort')  # добавляем подпись
            .background_gradient(cmap='viridis')  # раскрашиваем ячейки по столбцам
            .highlight_null('white')  # делаем белый фон для значений NaN
            .format("{:.2%}", na_rep=""))  # числа форматируем как проценты, NaN заменяем на пустоту
ur_style
```
